## Step #6:  The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:

## If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
## If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”


```python
# Create a new column, 'avg_order', for the average price of each customer's orders.
```

```python
ords_prods_merge['avg_order'] = ords_prods_merge.groupby(['customer_id'])['prices'].transform(np.mean)
```

```python
# Basic sanity-checking with the 'shape' and 'head()' functions.
```

```python
ords_prods_merge.shape
```

```python
ords_prods_merge.head(2)
```

```python
# Set up the 'spending_flag' column with our conditional statements.
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_order'] > 10, 'spending_flag'] = 'High spender'
ords_prods_merge.loc[ords_prods_merge['avg_order'] <= 10, 'spending_flag'] = 'Low spender'
```

```python
# Basic sanity-checking with the 'value_counts' and 'head()' functions.
```

```python
ords_prods_merge['spending_flag'].value_counts()
```

```python
# We have to look beyond the first 10 results to see a 'High spender'.
```

## Step #7:  In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:

## If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
## If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
## If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”


```python
# Create a new column, 'avg_order_frequency', for the average frequency of each customer's orders.
```

```python
ords_prods_merge['avg_order_frequency'] = ords_prods_merge.groupby(['customer_id'])['days_since_prior_order'].transform(np.mean)
```

```python
# Basic sanity-checking with the 'shape' and 'head()' functions.
```

```python
ords_prods_merge.shape
```

```python
ords_prods_merge.head(2)
```

```python
# Set up the 'frequency_flag' column with our conditional statements.
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_order_frequency'] > 20, 'frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['avg_order_frequency'] <= 20) & (ords_prods_merge['avg_order_frequency'] > 10), 'frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['avg_order_frequency'] <= 10, 'frequency_flag'] = 'Frequent customer'
# Extra line of comments, to make selecting on the bottomm line easier in Jupyter.  Otherwise, the scroll bar is a hindrance.
```

```python
# Basic sanity-checking with the 'value_counts' and 'head()' functions.
```

```python
ords_prods_merge['frequency_flag'].value_counts()
```

```python
ords_prods_merge[['customer_id', 'frequency_flag', 'avg_order_frequency']].head(60)
```
